```python
# Cargamos la session
import findspark
findspark.init()

from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
# Creamos variable de sparkContext
sc = spark.sparkContext
```

```python
# Crear RDD vacio (Sin particion)
rdd_vacio = sc.emptyRDD
```

```python
# Crear RDD vacio (con 3 particiones)
rdd_vacio3 = sc.parallelize([], 3)
rdd_vacio3.getNumPartitions()
```

```python
# Crear RDD con datos
rdd = sc.parallelize([1, 2, 3, 4, 5])
rdd
```

```python
rdd.collect()
```

```python
# Crear RDD desde un archivo de texto
rdd_texto = sc.textFile("./rdd_source.txt")
rdd_texto.collect()
```

```python
# Crear RDD donde todo el exto es un solo registro
rdd_texto_completo = sc.wholeTextFiles("./rdd_source.txt")
rdd_texto_completo.collect()
```

```python
rdd_suma = rdd.map(lambda x: x + 1)
rdd_suma.collect()
```

```python
# Pasar DF a RDD
df = spark.createDataFrame([(1, "José"),(2, "Juan")], ["id", "nombre"])
df.show()
```

```python
rdd_df = df.rdd
rdd_df.collect()
```
