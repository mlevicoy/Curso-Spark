**Agregación con agrupación**

```python
vuelos = spark.read.parquet("./data64/")
vuelos.printSchema()
```

```python
from pyspark.sql.functions import desc
```

```python
# Agrupamos por el aeropuerto de origen, contar cuantos vuelos hay desde el aeropuerto
# de origen y ordenarlo descendiente
(vuelos.groupBy('ORIGIN_AIRPORT')
    .count()
    .orderBy(desc('count'))
).show()
```

**Agregación con pivote**

```python
estudiantes = spark.read.parquet("./data66/")
```

```python
estudiantes.show()
```

```python
from pyspark.sql.functions import min, max, avg, col
```

```python
# Promedio por año de graduación
estudiantes.groupBy('graduacion').pivot('sexo').agg(avg('peso'), min('peso'), max('peso')).show()
```

```python
# Lo mismo, pero con el peso mínimo y máximo
estudiantes.groupBy('graduacion').pivot('sexo').agg(avg('peso')).show()
```

```python
# Lo mismo, pero solo hombres
estudiantes.groupBy('graduacion').pivot('sexo', ['M']).agg(avg('peso'), min('peso'), max('peso')).show()
```

```python
# Lo mismo, pero solo mujeres
estudiantes.groupBy('graduacion').pivot('sexo', ['F']).agg(avg('peso'), min('peso'), max('peso')).show()
```

**Inner Join**

```python
# El conjunto despues del JOIN contendra filas solo cuando la expresión se evalue como verdadera,
# es decir, cuando los valores de las columnas sean los mismos en ambos conjuntos.
```

```python
empleados = spark.read.parquet("./data69/empleados/")
departamentos = spark.read.parquet("./data69/departamentos/")
```

```python
empleados.show()
```

```python
departamentos.show()
```

```python
from pyspark.sql.functions import col
join_df = empleados.join(departamentos, col('num_dpto') == col('id'))
join_df.show()
```

```python
# otra forma
join_df = empleados.join(departamentos, col('num_dpto') == col('id'), 'inner')
join_df.show()
```

```python
# otra forma
join_df = empleados.join(departamentos).where(col('num_dpto') == col('id'))
join_df.show()
```

1. Determine los tres países con mayor número de jugadores(jugadores nacidos en ese país). El resultado debe estar ordenado de forma descendente.

```python
jugadores.filter(col('country_of_birth').isNotNull()).groupBy('country_of_birth').count().orderBy(desc('count')).limit(3).show()
```

2. Obtenga la lista de jugadores con tarjeta roja. La salida debe contener dos columnas, el nombre de pila del jugador y la cantidad de tarjetas rojas que tiene.

```python
df1 = jugadores.join(apariciones, ['player_id'], 'inner')
df2 = df1.filter(col('red_cards') > 0).select(col('pretty_name'), col('red_cards'))
df2.groupBy('pretty_name').count().orderBy(desc('count')).show()
```

3. ¿Cuántos juegos se jugaron en la Premier League? La salida debe contener dos columnas, el nombre de la liga y la cantidad de juegos que se jugaron en ella.

```python
data1 = juegos.join(competiciones, col('competition_id') == col('competition_code'), 'left')
```

```python
data1.show()
```

```python
data1.groupBy('name').count().filter(col('name') == 'premier-league').show()
```

4. Obtenga las tres ligas con mayor número de asistencia de público teniendo en cuenta todos los juegos que se jugaron en ellas. El resultado debe estar ordenado de forma descendente y tener dos columnas, el nombre de la liga y la asistencia total.

```python
from pyspark.sql.functions import sum
```

```python
data1.groupBy('name').agg(sum('attendance').alias('asistencia')).orderBy(desc('asistencia')).show()
```
