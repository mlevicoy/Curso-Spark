**Generamos la session**

```python
import findspark
findspark.init()

from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

**Generamos el contexto**

```python
sc = spark.sparkContext
```

**Función reduce**

```python
rdd = sc.parallelize([2, 4, 6, 8])
rdd.reduce(lambda x,y: x + y)
```

```python
rdd1 = sc.parallelize([1,2,3,4])
rdd1.reduce(lambda x,y: x * y)
```

**Función count**

```python
rdd = sc.parallelize(['j', 'o', 's', 'e'])
```

```python
rdd.count()
```

```python
rdd1 = sc.parallelize([item for item in range(10)])
```

```python
rdd1.count()
```

**Función collect**: Recopila todos los elementos de RDD y los envía al controlador

```python
rdd = sc.parallelize('Hola Apache Spark!'.split(' '))
```

```python
rdd.collect()
```

```python
rdd1 = sc.parallelize([(item, item ** 2) for item in range(20)])
```

```python
rdd1.collect()
```

**Función take, max y saveAsTextFile**

* take: toma n y devuelve los primero n elementos.
* max: Devuelve el máximo número.
* saveTextFile: Guarda en un archivo.

```python
rdd = sc.parallelize('La programación es bella'.split(' '))
```

```python
rdd.take(2)
```

```python
rdd1 = sc.parallelize([item/(item + 1) for item in range(10)])
```

```python
rdd1.max()
```

```python
rdd.saveAsTextFile('./rdd')
```

```python
rdd1.coalesce(1).saveAsTextFile('./rdd1')
```

1. Cree un RDD llamado importes a partir del archivo adjunto a esta lección como recurso.

1.c. Cree un RDD top15 que contenga los 15 mayores valores del RDD importes. Tenga en cuenta que pueden repetirse los valores. Por último, escriba el RDD top15 como archivo de texto en la carpeta data/salida.

```python
top15 = importes.sortBy(lambda x: -x)
top15.take(15)
```

```python
top15.coalesce(1).saveAsTextFile("./data/salida1")
```

```python
lista_top15 = importes.top(15)
top15 = sc.parallelize(lista_top15)
```

2. Cree una función llamada factorial que calcule el factorial de un número dado como parámetro. Utilice RDDs para el cálculo.

```python
import math

def factorial(c):
  return math.factorial(c)

rdd_factorial = sc.parallelize([(item, factorial(item)) for item in range(10)])
rdd_factorial.collect()
```

```python
def factorial_sin_funcion(n):
  if n == 0:
    return 1
  else:
    rdd = sc.parallelize(list(range(1, n+1)))
    return rdd.reduce(lambda x,y: x * y)

factorial_sin_funcion(4)
```
